```python
def g():
    def classify(x1, x2):
        return x1 % 2
            
    X = []
    y = []
    for col, i in enumerate(np.arange(0, 10, 10/32, dtype=float)):
        for j in np.arange(0, 10, 10/32, dtype=float):
            X.append(np.array([i, j]))
            y.append(col % 2)
    X = np.asarray(X)
    y = np.asarray(y)
    return X, y
    
X, y = g()
plot_2d(X, y)
```

lr > lda

```python
def g_lda():
    X1 = np.concatenate(
        (
            np.random.normal(4.8, .2, size=(150, 1)), 
            np.random.normal(5.5, 1.1, size=(150, 1))
        ),
        axis=1
    )
    y1 = np.zeros((X1.shape[0],))
    
    X2 = np.concatenate(
        (
            np.random.normal(7.1, 1.1, size=(150, 1)), 
            np.random.normal(4.2, 2.1, size=(150, 1))
        ),
        axis=1
    )
    y2 = np.ones((X2.shape[0],))
    
    X = np.concatenate((X1, X2,), axis=0)
    y = np.concatenate((y1, y2,), axis=0)
    return X, y
    
X, y = g_lda()
plot_2d(X, y)
```

```python
lr = LogisticRegression(max_iter=1000)
test_model(lr, X, y)
```

```python
lda = LinearDiscriminantAnalysis()
test_model(lda, X, y)
```

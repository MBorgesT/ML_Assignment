```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib import pyplot as plt
```

```python
def G_SVC_linear(n):

    X,y = datasets.make_classification(n_samples=n,n_features=2,
                                       n_clusters_per_class=2,
                                       n_redundant=0,
                                       class_sep=2,
                                       random_state=42)
    
    return X, y
```

```python
list_linear = {}
list_rbf = {}



for n in range(100, 10000, 500):
    
   
    
    X_lin, y_lin = G_SVC_linear(n)
    X_train, X_test, y_train, y_test = train_test_split(X_lin, y_lin, test_size=.33, random_state=42)
    model_linear = SVC(kernel='linear', random_state=42).fit(X_train, y_train)
    model_rbf = SVC(kernel='rbf', random_state=42).fit(X_train, y_train)

    list_linear[n] =(model_linear.score(X_test, y_test) * 100)
    list_rbf[n] = (model_rbf.score(X_test, y_test) * 100)
    
    plt.scatter(X_lin[:, 0], X_lin[:, 1], c = y_lin)

        
        
        #list_linear.append((model_linear.score(X_test, y_test) * 100))
        #list_rbf.append((model_rbf.score(X_test, y_test) * 100))



```

```python
plt.plot(list_linear.keys(), list_linear.values(), label='linear')
plt.plot(list_rbf.keys(), list_rbf.values(), label='rbf')
plt.legend()
```
